```python
import pandas as pd
```

```python
variable_dict_df = pd.read_csv('Data/American Community Survey Data/acs5_variable_dict_2014_2019.csv')
```

```python
variable_dict_df.head()
```

```python
variable_dict_df.dtypes
```

```python
#Count the No. of variable codes in 2014 and 2019
variable_dict_df.groupby('acs_year')['variable_code'].count()
```

There are more variable codes in 2019 than in 2014. To be exact, there are 88 extra codes in 2019

```python
# Check what variable_codes appear in 2014 set alone or 2019 set alone
variable_dict_2014 = variable_dict_df.loc[variable_dict_df['acs_year']==2014]
variable_dict_2019 = variable_dict_df.loc[variable_dict_df['acs_year']==2019]
```

```python
# Find variable codes that appear in 2014 but not in 2019
variable_code_2019 = variable_dict_2019['variable_code'].tolist()
for variable_code in variable_dict_2014['variable_code']:
    if variable_code not in variable_code_2019:
        print(variable_code)
```

All varible codes that appear in 2014 also appear in 2019

```python
# Find variable codes that appear in 2019 but not in 2014
count_codes = 0
variable_code_2014 = variable_dict_2014['variable_code'].tolist()
for variable_code in variable_dict_2019['variable_code']:
    if variable_code not in variable_code_2014:
        count_codes +=1
        print(variable_code)
        
print ('The No of variable codes that appear in 2019 but not in 2014: {}'.format(count_codes))
```

The No of variable codes that appear in 2019 but not in 2014: 88

```python
#count the variable codes whose labels are different between 2014 and 2019
count_mismatched_labels = 0        
for var_code in variable_code_2014:
    label_2014_series = variable_dict_2014.loc[variable_dict_2014['variable_code']==var_code]['label']
    label_2014 = label_2014_series[label_2014_series.index[0]] #since only one entry is available for each variable code in agiven year
    label_2019_series = variable_dict_2019.loc[variable_dict_2019['variable_code']==var_code]['label']
    label_2019 = label_2019_series[label_2019_series.index[0]] #since only one entry is available for each variable code in agiven year
    if  label_2014 !=  label_2019:
        count_mismatched_labels +=1
        print(var_code)
```

```python
print('No of variable codes whose labels are different in 2014 and 2019 : {}'.format(count_mismatched_labels))
```

Out of the 4088 common variable codes between 2014 and 2019, 86.69% of the variable codes didn't have the same label for both years. This is based on an exact string comparison.
