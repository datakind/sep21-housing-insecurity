```python
orange_2014 = pd.read_csv("/home/audry/Documents/datakind/orange_acs5-2014_census.csv")
orange_2019 = pd.read_csv("/home/audry/Documents/datakind/orange_acs5-2019_census.csv")
```

```python
orange_2014col = list(orange_2014.columns)
orange_2019col = list(orange_2019.columns)
```

```python
orange_2014col2 = ["C2014_" + i for i in orange_2014col]
orange_2019col2 = ["C2019_" +i for i in orange_2019col]
```

```python
orange_2014.columns = orange_2014col2
orange_2019.columns = orange_2019col2
```

```python
orange_2014.head()
```

```python
bothyears = pd.concat([orange_2014,orange_2019],axis = 1)
```

```python
bothyears.shape
```

```python
orange_2014.shape
```

```python
from scipy import stats
```

```python
#Goal is to run a paired t-test on those that have the same variable names, with the caveat that variable names are not necessarily the answer. 
failreason = pd.DataFrame()
sigtest = pd.DataFrame()
for i in orange_2014col:
    #Check to see if label's match
    tmp = df[df.variable_code==i]
    failcheck = 0
    if tmp.shape[0] != 2:
        tmpfail = pd.DataFrame({"failreason":"not in both","Var_name": i,"Var_label":"None"},index =[0])
        failreason=failreason.append(tmpfail,ignore_index = True)
        failcheck = 1
        continue
    if tmp.label.iloc[0] != tmp.label.iloc[1]:
        tmpfail = pd.DataFrame({"failreason":"label_mismatch","Var_name": i,"Var_label":tmp.label.iloc[0]},index =[0])
        failreason=failreason.append(tmpfail, ignore_index = True)
        tmpfail = pd.DataFrame({"failreason":"label_mismatch","Var_name": i,"Var_label":tmp.label.iloc[1]},index =[0])
        failreason=failreason.append(tmpfail, ignore_index = True)
        failcheck = 1
    if failcheck == 0:
        old = "C2014_"+i
        new = "C2019_"+i
        pvalues = stats.ttest_rel(bothyears[old], bothyears[new]).pvalue
        data = pd.DataFrame({"Var_label":tmp.label.iloc[0],"Var_name":i,"pvalue":pvalues,"failcheck":failcheck},index = [0])
        sigtest = sigtest.append(data, ignore_index = True) #This is a paired t-test 
```

```python
sigtest = sigtest.sort_values(by = "pvalue",ascending=True)
sigtest[0:10] #lowest p-value from paired t-test
```
